```python
# goldstone_hyper_holo_fractal_boosted_fixed.py
# QCD-inspired SSB/Goldstone kernel with fractal-holo resolution
# Lorenz task, upgraded coherence metrics, robust PCA viz

import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

class GoldstoneHyperHoloBlock(nn.Module):
    def __init__(self, dim=128, depth=5, alpha=1.7, slip_factor=0.15, ng_strength=0.08):
        super().__init__()
        self.depth = depth
        self.alpha = alpha
        self.proj = nn.Linear(dim, dim)
        self.slip_gen = nn.Linear(dim, dim)
        self.holo_proj = nn.Linear(dim * 2, dim * 4)
        self.inv_holo = nn.Linear(dim * 4, dim)
        self.ng_proj = nn.Linear(dim, dim)
        self.ng_strength = ng_strength
        self.slip_factor = slip_factor
        self.norm_h = nn.LayerNorm(dim)
        self.norm_out = nn.LayerNorm(dim)

    def forward(self, x, level=0, t_step=0.0, **kwargs):
        if level >= self.depth:
            return x

        slip = torch.tanh(self.slip_gen(x)) * self.slip_factor
        h = torch.relu(self.proj(x + slip))
        h = self.norm_h(h)

        h_child = self.forward(h, level + 1, t_step)

        concat = torch.cat([h, h_child], dim=-1)
        boundary = self.holo_proj(concat)
        resolved = self.inv_holo(boundary)
        fb = resolved * h

        vev_rand = torch.tanh(torch.randn_like(h.mean(dim=-1, keepdim=True)))
        vev = torch.norm(h, dim=-1, keepdim=True) * 0.005 * vev_rand
        h_broken = h + vev

        proj = self.ng_proj(h_broken)
        norm_hb_sq = h_broken.norm(dim=-1, keepdim=True)**2 + 1e-6
        dot = (proj * h_broken).sum(-1, keepdim=True) / norm_hb_sq
        goldstone = proj - dot * h_broken
        phase = torch.sin(t_step * 2.0 * np.pi + torch.randn_like(goldstone) * 0.2)

        h = h + self.ng_strength * goldstone * phase
        h = self.norm_h(h)

        out = x + fb + 0.05 * h_child
        mask_prob = max(0.1, 1.0 / ((level + 1) ** self.alpha))
        mask = (torch.rand_like(out) < mask_prob).float()
        out = out * mask
        out = self.norm_out(out)

        return out

class CoherenceNet(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=128):
        super().__init__()
        self.embed = nn.Linear(input_dim, hidden_dim)
        self.fractal_core = GoldstoneHyperHoloBlock(hidden_dim)
        self.head = nn.Linear(hidden_dim, input_dim)

    def forward(self, x, steps=200):
        h = self.embed(x)
        states = [h.clone()]
        for t in range(steps):
            h = self.fractal_core(h, t_step=t / steps)
            if torch.isnan(h).any() or torch.isinf(h).any():
                print(f"NaN/Inf in h at step {t}")
                break
            states.append(h.clone())
        out = self.head(h)
        return out, torch.stack(states)

def generate_noisy_lorenz(n_steps=2000, dt=0.01, noise_std=0.05):
    def lorenz_step(xyz, s=10, r=28, b=8/3):
        x, y, z = xyz
        return np.array([s*(y - x), x*(r - z) - y, x*y - b*z])

    states = np.zeros((n_steps + 1, 3))
    states[0] = [0.1, 0.0, 0.0]
    for i in range(n_steps):
        deriv = lorenz_step(states[i])
        states[i+1] = states[i] + dt * deriv + noise_std * np.random.randn(3)
    return torch.tensor(states, dtype=torch.float32)

def test_run(steps=200, epochs=100):
    model = CoherenceNet(input_dim=3)
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
    criterion = nn.MSELoss()

    data = generate_noisy_lorenz(n_steps=steps + 10)  # Slightly longer than rollout
    x_init = data[0:1, :]   # [1, 3] initial condition (batch=1)

    for epoch in range(epochs):
        out, states = model(x_init, steps=steps)
        # Target: actual next steps from data (first rollout step for simplicity)
        target = data[1:steps+1, :]  # [steps, 3]

        # Compare only the final prediction to the corresponding target step
        loss = criterion(out, target[-1:])  # or average over rollout if you extend out

        if torch.isnan(loss).any() or torch.isinf(loss).any():
            print(f"NaN/Inf loss at epoch {epoch}")
            break

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        if epoch % 10 == 0:
            print(f"Epoch {epoch}: Loss {loss.item():.4f}")
            norm_std = states.norm(dim=-1).std()
            print(f"  Norm coherence (std over {steps} steps): {norm_std.item():.4f}")
            cos_sims = torch.nn.functional.cosine_similarity(states[:-1], states[1:], dim=-1).mean()
            print(f"  Avg cosine sim (consecutive states): {cos_sims.item():.4f}")
            state_var = states.var(dim=-1).mean()
            print(f"  Avg state variance: {state_var.item():.4f}")
            final_norm = states[-1].norm(dim=-1).mean()
            print(f"  Final state avg norm: {final_norm.item():.4f}")

    # Robust PCA viz
    states_np = states.detach().cpu().numpy()  # [steps+1, 1, hidden_dim] or [steps+1, hidden_dim]
    if len(states_np.shape) == 3:
        states_flat = states_np.reshape(-1, states_np.shape[-1])  # Flatten batch & steps
    else:
        states_flat = states_np

    print(f"PCA input shape: {states_flat.shape}")

    pca = PCA(n_components=2)
    proj = pca.fit_transform(states_flat)

    plt.figure(figsize=(8, 6))
    plt.plot(proj[:, 0], proj[:, 1], '-o', markersize=3, alpha=0.7)
    plt.title("PCA Trajectory of Hidden States (Lorenz Task, 200 Steps)")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid(True)
    plt.savefig("attractor_trajectory.png")
    plt.show()
    print(f"PCA explained variance ratio: {pca.explained_variance_ratio_}")

    return model

if __name__ == "__main__":
    print("Sweeping the leg: Running boosted kernel on Lorenz...")
    test_run()
    print("Kernel executed. Bootstrapping checked — physics funding inbound.")
```
